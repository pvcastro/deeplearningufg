```python
import re
import sys

def padronizar_arquivo(folder, file_in, file_out, total=False):
    
    cenario_total = ['PESSOA', 'ORGANIZACAO', 'LOCAL', 'VALOR', 'TEMPO', 'ABSTRACCAO', 'OBRA', 'ACONTECIMENTO', 'COISA', 'OUTRO']
    cenario_seletivo = ['PESSOA', 'ORGANIZACAO', 'LOCAL', 'VALOR', 'TEMPO']
    
    f = open(folder + file_in, 'r')
    f_str = f.read().splitlines()
    f.close()
    lines = []

    for i,line in enumerate(f_str[1:]):
        if not line.startswith('#') > 0:
            if len(line):
                tokens = line.split()
                palavra = tokens[0]
                tag = tokens[2]
                tokens_tag = tag.split('-')
                if len(tokens_tag) > 1 and total is False and tokens_tag[1] not in cenario_seletivo:
                    tag = 'O'
                lines.append(palavra + '\t' + tag)
            else:
                lines.append('')

    to_file = "\n".join(lines)

    if total is True:
        prefixo = 'categories'
    else:
        prefixo = 'filtered'
    
    o = open(folder + prefixo + file_out,'w')
    o.write(to_file)
    o.close()
```

```python
folder = '/media/discoD/Mestrado/Corpora/Cicero/harem_dataset/'

padronizar_arquivo(folder=folder, file_in='harem.train.txt', file_out='_harem_I_cicero.txt', total=True)
padronizar_arquivo(folder=folder, file_in='harem.train.txt', file_out='_harem_I_cicero.txt', total=False)
padronizar_arquivo(folder=folder, file_in='MiniHarem.txt', file_out='_mini_cicero.txt', total=True)
padronizar_arquivo(folder=folder, file_in='MiniHarem.txt', file_out='_mini_cicero.txt', total=False)
```
