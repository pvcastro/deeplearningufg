```python


import pandas as pd
from sklearn.model_selection import train_test_split
import html, re
```

```python
re1 = re.compile(r'  +')

def fixup(x, title=False):
    x = x.replace('#39;', "'").replace('amp;', '&').replace('#146;', "'").replace(
        'nbsp;', ' ').replace('#36;', '$').replace('\\n', "\n").replace('quot;', "'").replace(
        '<br />', "\n").replace('\\"', '"').replace('<unk>','u_n').replace(' @.@ ','.').replace(
        ' @-@ ','-').replace('\\', ' \\ ')
    if title:
        x = x.replace('\n', '').replace('\t', '')
    return re1.sub(' ', html.unescape(x))

def concat_text_title(row):
    separator = ' ' if row['title'].endswith('.') else '. '
    return row['title'] + separator + row['text']
```

```python
news_df = pd.read_csv('/media/discoD/Mestrado/FASAM/df_train.csv')
news_df['title'] = news_df.apply(lambda row: fixup(row['title'], title=True), axis=1)
news_df['text'] = news_df.apply(lambda row: fixup(row['text'], title=False), axis=1)
news_df['text'] = news_df.apply(lambda row: concat_text_title(row), axis=1)
news_df['label'] = news_df['category']
del news_df['category']
del news_df['title']
news_df = news_df.reindex(columns = ['label','text'])
news_df.to_json('/media/discoD/Mestrado/FASAM/df_train.jsonl', orient='records', lines=True)
news_df
```
