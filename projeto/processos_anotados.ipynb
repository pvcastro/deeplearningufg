```python
def drop_duplicate_sort(filename, name):
    df = pd.read_csv(filename, header=None, sep='\t')
    df = df.rename(columns = {0:name})
    df[name].sort_values(ascending=True)
    df[name.lower()] = df.apply(lambda x: x[name].lower(), axis=1)
    df = df.drop_duplicates(name.lower())
    df = df.sort_values(by=[name.lower()], ascending=True)
    df.to_csv(filename.replace('.csv', '_sorted.csv'), columns=[name], index=None)
    return df
```

```python
funcoes_df = drop_duplicate_sort('funcoes.csv', 'Funo')
funcoes_df
```

```python
fundamentos_df = drop_duplicate_sort('fundamentos.csv', 'Fundamento')
fundamentos_df
```
