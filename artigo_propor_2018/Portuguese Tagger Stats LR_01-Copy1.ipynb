```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

pd.options.display.max_columns = None
pd.options.display.max_rows = None
pd.options.display.float_format = '{:,.2f}'.format
```

```python
def agrupar_colunas(row, colunas):
    valores = []
    for coluna in colunas:
        valores.append(str(row[coluna]))
    return ', '.join(valores)
```

```python
def outliers_iqr(values):
    quartile_1, quartile_3 = np.percentile(values, [50, 75])
    print('quartile 1, quartile 3', quartile_1, quartile_3)
    iqr = quartile_3 - quartile_1
    print('iqr', iqr)
    lower_bound = quartile_1 - (iqr * 1.5)
    print(lower_bound)
    upper_bound = quartile_3 + (iqr * 1.5)
    print(upper_bound)
    #return np.where((values > upper_bound) | (values < lower_bound))
    result = (values <= upper_bound) & (values >= lower_bound)
    #print(result)
    return result
```

```python
dados_df = pd.read_csv('resultados.csv')
#dados_df['Short Embedding'] = dados_df.apply(lambda row: get_short_embedding(row['Embedding']), axis=1)
dados_df
```

```python
dados_df['Capitalization / Normalization'] = dados_df.apply(lambda row: agrupar_colunas(row, colunas=['Capitalisation Dim','Lower']), axis=1)
```

```python
dados_df['Hidden Units'] = dados_df.apply(lambda row: agrupar_colunas(row, colunas=['Char LSTM Dim','Word LSTM Dim']), axis=1)
```

```python
dados_df
```

```python
groups = dados_df.groupby(['Epochs','Embedding','Scenario','Tagging Scheme','Capitalization / Normalization','Hidden Units'])
```

```python
len(groups)
```

```python
first_group = groups.get_group((5,'Wang2Vec','selective','iob','False, True', '50, 200'))
first_group
```

```python
outliers_iqr(first_group['F1 Score'])
```

```python
filtered_groups = []
for group in groups.groups:
    group_df = groups.get_group(group)
    #print(group_df)
    group_df = group_df[outliers_iqr(group_df['F1 Score'])]
    #print(group_df)
    #print(group_df['F1 Score'], outliers_iqr(group_df['F1 Score']))
    #filtered = outliers_iqr(group_df['F1 Score'].tolist())
    #print(filtered[0], type(filtered))
    #filtered = filtered.tolist()
    #print(filtered, type(filtered))
    #filter_list += filtered[0]
    filtered_groups.append(group_df)
```

```python
dados_df = pd.concat(filtered_groups)
dados_df
```

```python
def plot_and_save(data_frame, column, title, prefix='plt', save=False):
    data_frame.boxplot(column='F1 Score', by=column, showmeans=True, fontsize=16, showfliers=False)
    plt.title(title, fontsize=18)
    plt.suptitle("")
    plt.xlabel("")
    if save == True:
        filename = column.replace(' / ', '_').replace(' ','_').lower()
        plt.savefig(prefix + "_" + filename + ".pdf", bbox_inches='tight')
```

```python
plot_and_save(dados_fscore_df, 'Embedding', 'Embeddings')
plot_and_save(dados_fscore_df, 'Tagging Scheme', 'Tagging Schemes')
plot_and_save(dados_fscore_df, 'Capitalization / Normalization', 'Capitalization / Normalization')
plot_and_save(dados_fscore_df, 'Hidden Units', 'Hidden Units')
plot_and_save(dados_df, 'Lower', 'Normalization')
plot_and_save(dados_df, 'Capitalisation Dim', 'Capitalization')
plot_and_save(dados_df, 'Char LSTM Dim', 'Character LSTM Units')
plot_and_save(dados_df, 'Word LSTM Dim', 'Word LSTM Units')
```
