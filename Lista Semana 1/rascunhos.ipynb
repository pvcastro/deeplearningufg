```python
from random import choice 
from numpy import array, dot, random 
import numpy as np

unit_step = lambda x: 0 if x < 0 else 1 

training_data = [ (array([0,0,1]), 0), (array([0,1,1]), 1), (array([1,0,1]), 1), (array([1,1,1]), 1), ] 

w = np.zeros(3)
errors = [] 
eta = 0.1
n = 100 

for i in range(n): 
    print(w)
    x, expected = choice(training_data) 
    print("x: " + str(x))
    print("expected: " + str(expected))
    result = dot(w, x) 
    print("result: " + str(result))
    error = expected - unit_step(result) 
    print("error: " + str(error))
    errors.append(error) 
    w += eta * error * x 
    
for x, _ in training_data: 
    result = dot(x, w) 
    print("{}: {} -> {}".format(x[:2], result, unit_step(result)))
```

```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt

class AdalineGD(object):

    def __init__(self, eta=0.01, epochs=50):
        self.eta = eta
        self.epochs = epochs

    def train(self, X, y):

        self.w_ = np.zeros(1 + X.shape[1])
        print("pesos: " + str(self.w_))
        self.cost_ = []

        for i in range(self.epochs):
            for i in range(len(X)):
                entrada = X[i]
                print("entrada: " + str(entrada))
                saida = y[i]
                print("saida: " + str(saida))
                output = self.net_input(entrada)
                print("output: " + str(output))
                errors = (saida - output)
                print("errors: " + str(errors))
                self.w_[1:] += self.eta * entrada * errors
                print("pesos atualizados: " + str(self.w_[1:]))
                self.w_[0] += self.eta * errors.sum()
                print("bias atualizado: " + str(self.w_[0]))
            cost = (errors**2).sum() / 2.0
            print("cost: " + str(self.cost_))
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        return self.net_input(X)

    def predict(self, X):
        return np.where(self.activation(X) >= 0.0, 1, -1)
```

```python
X = array([ array([0,0]), array([0,1]), array([1,0]), array([1,1]) ])
y = [ 0, 1, 1, 1 ] 

ada = AdalineGD(epochs=500, eta=0.1).train(X, y)
plt.plot(range(1, len(ada.cost_)+1), np.log10(ada.cost_), marker='o')
plt.xlabel('Iterations')
plt.ylabel('log(Sum-squared-error)')
plt.title('Adaline - Learning rate 0.01')
plt.show()

ada = AdalineGD(epochs=10, eta=0.0001).train(X, y)
plt.plot(range(1, len(ada.cost_)+1), ada.cost_, marker='o')
plt.xlabel('Iterations')
plt.ylabel('Sum-squared-error')
plt.title('Adaline - Learning rate 0.0001')
plt.show()
```
